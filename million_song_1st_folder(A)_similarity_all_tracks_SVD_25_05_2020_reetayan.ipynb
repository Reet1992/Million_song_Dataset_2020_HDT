```python
#### Singular Value Decomposition
```

```python
A = pivot_A_Final.as_matrix()
track_ratings_mean = np.mean(A, axis = 1)
A_demeaned = A - track_ratings_mean.reshape(-1, 1)
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(A_demeaned, k = 7)
```

```python
pd.DataFrame(U)
```

```python
pd.DataFrame(sigma)
```

```python
pd.DataFrame(Vt)
```

```python
sigma = np.diag(sigma)
```

```python
sigma
```

```python
all_track_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + track_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_track_predicted_ratings, columns = pivot_A_Final.columns)
```
